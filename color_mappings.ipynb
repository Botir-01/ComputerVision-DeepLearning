```python
import cv2 
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/00-puppy.jpg')
```

```python
plt.imshow(image)
```

```python
fixed_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(fixed_image)
```

```python
HSL_image = cv2.cvtColor(fixed_image, cv2.COLOR_BGR2HLS)
```

```python
plt.imshow(HSL_image)
```

```python
HSV_image = cv2.cvtColor(fixed_image, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(HSV_image)
```
